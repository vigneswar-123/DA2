```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.DataFrame({'Order_time':[12.00,12.20,12.30,17.00,20.00,20.20,20.30,20.10],
                 'delivery_time':[12.10,12.30,12.40,17.40,20.20,20.40,20.55,20.30],
                 'delivery_partner':["swiggy","zomato","swiggy","swiggy","swiggy","zomato","zomato","swiggy"],
                 'customer_ratings':[5,4,4,2,3,3,2,3]
                 })
df.head()
```

```
avg_del=df.groupby('delivery_partner')['delivery_time'].mean()
print(avg_del)
```

```
peak = df['Order_time'].astype(int).mode()
print(f"The peak order hour is: {peak.iloc[0]}")
```

```
df["delay"]=df['Order_time']-df["delivery_time"]
df.head()
```

```
most_delayed_partner = df.loc[df["delay"].idxmin(), 'delivery_partner']
print(f"The delivery partner with the most delay is: {most_delayed_partner}")
```

```
most_delayed_partner = df.loc[df["delay"].idxmax(), 'delivery_partner']
print(f"The delivery partner with the fastest delivery is: {most_delayed_partner}")
```

```
most_delayed_partner = df.loc[df["customer_ratings"].idxmin(), 'delivery_partner']
print(f"The delivery partner with least rating is: {most_delayed_partner}")
```

```
most_delayed_partner = df.loc[df["customer_ratings"].idxmax(), 'delivery_partner']
print(f"The delivery partner with most rating is: {most_delayed_partner}")
```

```
no_of_deli=df['delivery_partner'].value_counts()
print(no_of_deli)
```

```
plt.scatter(df['delivery_time'],df['Order_time'])
plt.xlabel('delivery_time')
plt.ylabel('Order_time')
plt.title('Order_time vs delivery_time')
plt.show()
```

```
plt.bar(df['delivery_partner'],df['customer_ratings'])
plt.xlabel('delivery_partner')
plt.ylabel('customer_ratings')
plt.title('delivery_partner vs customer_ratings')
plt.show()
```

```
plt.bar(df['delivery_time'],df['customer_ratings'])
plt.xlabel('delivery_time')
plt.ylabel('customer_ratings')
plt.title('delivery_time vs customer_ratings')
plt.show()
```
