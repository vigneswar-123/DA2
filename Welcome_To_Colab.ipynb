```
#Smart Water Tank
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.DataFrame({'day':["mon","tue","wed","thu","fri","sat","sun"],
                 'water_con':[200,170,180,195,175,210,220]})
tank_capacity=220
df['water_req']=tank_capacity-df['water_con']
print(df)
```

```
mean_con=df['water_con'].mean()
print(mean_con)
```

```
total_difference = df['water_con'].diff()
print(total_difference)
```

```
sum_total_diff=total_difference.sum()
print(sum_total_diff)
```

```
avg=sum_total_diff/7
print(avg)
```

```
new_cap=tank_capacity+avg
print(new_cap)
```

```
last_day_water_con=df['water_con'][6]
print(last_day_water_con)
```

```
if new_cap > last_day_water_con:
  print("water shortage will come.please! use limitedly")
else:
  print("no shortage!, use according to the needs")
```
