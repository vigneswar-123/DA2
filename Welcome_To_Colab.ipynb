```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.DataFrame({'cust_id':["ramesh","suresh","puja","raju","disha","manu","radhika","babu","ganesh"],
                 'internet_speed(mbps)':[30,20,60,40,50,15,10,70,60],
                 'payment_failure':["success","fail","success","success","success","fail","fail","success","success"],
                 "delivary_charges":[55,60,29,30,35,45,50,40,35],
                 "purchase_completed":["yes","no","yes","yes","yes","no","no","yes","yes"]
                 })
df.head()
```

```
df["del_avg"]=df['delivary_charges'].mean()
df["int_avg"]=df['internet_speed(mbps)'].mean()
df.head()
```

```
low_speed_count = (df['internet_speed(mbps)'] < 39.444).sum()
print(low_speed_count)
```

```
low_speed_count = df[df['internet_speed(mbps)'] < 39.444]["cust_id"].tolist()
print(low_speed_count)
```

```
high_del = df[df['delivary_charges'] > 41]['cust_id'].reset_index(drop=True)
print(high_del)
```

```
success_customers = df[df['payment_failure'] == 'success']['cust_id'].tolist()
print(success_customers)
```

```
fail_customers = df[df['payment_failure'] == 'fail']['cust_id'].tolist()
print(fail_customers)
```

```
result=df[(df['payment_failure'] == 'fail') &
    (df['delivary_charges'] > 42)&(df['internet_speed(mbps)']<39.5)]["cust_id"].reset_index(drop=True)
print(result)
```

```
result2=df[(df['payment_failure']=='success')&(df['delivary_charges']<42)&(df['internet_speed(mbps)']>39.5)]["cust_id"].reset_index(drop=True)
print(result2)
```

```
import matplotlib.pyplot as plt
churn_count=len(result)
loyal_count=len(result2)
print(churn_count)
print(loyal_count)
```

```
labels=['churn_count','loyal_count']
sizes=[churn_count,loyal_count]
colors=['red','green']
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%')
plt.title('churn vs loyal')
plt.show()
```
