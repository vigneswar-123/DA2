```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("weather_forecast_data.csv")
df.head()
```

```
last_temperature = df['Temperature'].iloc[-1]
print(last_temperature)
```

```
mean_temperature = df['Temperature'].mean()
print("mean_temperature",mean_temperature)
mean_humidity = df['Humidity'].mean()
print("mean_humidity",mean_humidity)
mean_wind_speed = df['Wind_Speed'].mean()
print("mean_wind_speed",mean_wind_speed)
mean_pressure = df['Pressure'].mean()
print("mean_pressure",mean_pressure)
```

```
diff_temperature = df['Temperature'].diff().sum()/7
print(diff_temperature)
```

```
if last_temperature>mean_temperature:
  print("Temperature is greater than mean temperature,so tmrw is also likely high temp")
else:
  print("Temperature is less than mean temperature,so tmrw is likely low temp")
```

```
last_humidity = df['Humidity'].iloc[-1]
print(last_humidity)
if last_humidity>mean_humidity:
  print("Humidity is greater than mean humidity,so tmrw is also likely high humidity")
else:
  print("Humidity is less than mean humidity,so tmrw is likely low humidity")
```

```
last_wind_speed = df['Wind_Speed'].iloc[-1]
print(last_wind_speed)
if last_wind_speed>mean_wind_speed:
  print("Wind_Speed is greater than mean wind_speed,so tmrw is also likely high wind_speed")
else:
  print("Wind_Speed is less than mean wind_speed,so tmrw is likely low wind_speed")
```

```
if last_humidity>mean_humidity:
  print("tmrw is maybe rainy")
  if last_wind_speed>mean_wind_speed:
    print("tmrw is most likely rainy")
    if last_temperature<mean_temperature:
      print("tmrw is rainy")
```
