```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("e_comerece.csv")
df.head()
```

```python
df.head()
```

```python
df['DaySinceLastOrder'] = df['DaySinceLastOrder'].fillna(0)
```

```python
display(df.head())
```

```python
df['Tenure']=df['Tenure'].fillna(df['Tenure'].mean())
```

```python
df['WarehouseToHome']=df['WarehouseToHome'].fillna(df['WarehouseToHome'].mean())
```

```python
mean_day=df['DaySinceLastOrder'].mean()
print(mean_day)
mean_ware=df['WarehouseToHome'].mean()
print(mean_ware)
mean_sat=df['SatisfactionScore'].mean()
print(mean_sat)
mean_ten=df['Tenure'].mean()
print(mean_ten)
mean_cash=df['CashbackAmount'].mean()
print(mean_cash)
```

```python
result1 = df[df['WarehouseToHome'] > mean_ware]['Churn'].value_counts()
print(result1)
```

```python
result2 = df[df['DaySinceLastOrder'] > mean_day]['Churn'].value_counts()
print(result2)
```

```python
result3 = df[df['SatisfactionScore'] > mean_sat]['Churn'].value_counts()
print(result3)
```

```python
result4 = df[df['Tenure'] > mean_ten]['Churn'].value_counts()
print(result4)
```

```python
result5= df[df['CashbackAmount'] > mean_cash]['Churn'].value_counts()
print(result5)
```

```python
result6=df[(df['WarehouseToHome'] > mean_ware)&(df['DaySinceLastOrder'] > mean_day)&(df['SatisfactionScore'] > mean_sat)&(df['Tenure'] > mean_ten)&(df['CashbackAmount'] > mean_cash)]['Churn'].value_counts()
print(result6)
```

```python
result7=df[(df['WarehouseToHome'] > mean_ware)&(df['DaySinceLastOrder'] > mean_day)]['Churn'].value_counts()
print(result7)
```

```python
result8=df[(df['Tenure'] > mean_ware)&(df['DaySinceLastOrder'] > mean_day)]['Churn'].value_counts()
print(result8)
```
