```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.DataFrame({'product_purchase':['laptop','mouse','webcam','keyboard','graphics_card'],
                 'sold_count':[25,30,10,20,5],
                 'cart_abandonment':[2,5,1,2,0],
                 'customer_ratings':[5,4,0,0,0],
                 'payment_methods':["cash","cash","online","online","online"],
                 'revenue':[90000,45000,35000,40000,15000]
                 })
df.head()
```

```python
max_sold_product_idx = df['sold_count'].idxmax()
max_pro = df.loc[max_sold_product_idx, 'product_purchase']
print(max_pro)
```

```python
most_revenue_idx = df['revenue'].idxmax()
most_rev = df.loc[most_revenue_idx, 'product_purchase']
print(most_rev)
```

```python
mode_pay=df['payment_methods'].mode()
print(mode_pay)
```

```python
high_rate=df['customer_ratings'].value_counts().idxmin()
print(high_rate)
```

```python
low_rate=df['customer_ratings'].value_counts().idxmax()
print(low_rate)
```

```python
purchased = df[df['cart_abandonment'] == 'false']
abandoned = df[df['cart_abandonment'] == 'true']
uncertain = df[df['cart_abandonment'] == 'maybe']
print(purchased)
print(abandoned)
print(uncertain)
```

```python
transactions = [
    ['laptop', 'mouse'],
    ['laptop', 'mouse', 'keyboard'],
    ['laptop'],
    ['laptop', 'mouse'],
    ['laptop', 'mouse', 'webcam'],
    ['mouse', 'keyboard'],
    ['laptop', 'mouse', 'graphics_card']
]

from collections import Counter

laptop_count = 0
co_occurrence = Counter()

for transaction in transactions:
    if 'laptop' in transaction:
        laptop_count += 1

        for product in transaction:
            if product != 'laptop':
                co_occurrence[product] += 1

print("Laptop appears in", laptop_count, "transactions\n")

for product, count in co_occurrence.items():
    confidence = (count / laptop_count) * 100
    print(f"Laptop -> {product}: {confidence:.2f}%")
```
